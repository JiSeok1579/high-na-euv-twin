# 3D Pupil Wavefront (Zernike)

Tier 1 matplotlib 3D surface for Zernike wavefront terms used by the pupil model.

```python
import os, sys
sys.path.insert(0, os.path.abspath('..'))

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401

from src.optics.zernike import wavefront
```

```python
grid = 160
axis = np.linspace(-1.0, 1.0, grid)
X, Y = np.meshgrid(axis, axis)
rho = np.sqrt(X**2 + Y**2)
theta = np.arctan2(Y, X)
coeffs = {(2, 0): 0.08, (2, 2): 0.04, (3, 1): -0.03}
W = wavefront(rho, theta, coeffs)
W = np.where(rho <= 1.0, W, np.nan)

fig = plt.figure(figsize=(7, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, W, cmap='coolwarm', linewidth=0, antialiased=True)
ax.set_xlabel('pupil x')
ax.set_ylabel('pupil y')
ax.set_zlabel('OPD [waves]')
ax.set_title('Zernike wavefront surface')
fig.tight_layout()
plt.show()
```

```python
# Parameter sweep: compare common Zernike modes used for qualitative tuning.
zernike_cases = [
    ('defocus', {(2, 0): 0.08}),
    ('astigmatism 0 deg', {(2, 2): 0.08}),
    ('astigmatism 45 deg', {(2, -2): 0.08}),
    ('coma x', {(3, 1): 0.08}),
    ('coma y', {(3, -1): 0.08}),
    ('primary spherical', {(4, 0): 0.08}),
]

fig = plt.figure(figsize=(13, 7))
for idx, (label, coeff) in enumerate(zernike_cases, start=1):
    surface = wavefront(rho, theta, coeff)
    surface = np.where(rho <= 1.0, surface, np.nan)
    ax = fig.add_subplot(2, 3, idx, projection='3d')
    ax.plot_surface(X, Y, surface, cmap='coolwarm', linewidth=0, antialiased=True)
    ax.set_title(label)
    ax.set_xlabel('pupil x')
    ax.set_ylabel('pupil y')
    ax.set_zlabel('OPD [waves]')
    ax.set_zlim(-0.1, 0.1)

fig.suptitle('Zernike mode comparison at 0.08 waves', y=0.98)
fig.tight_layout()
plt.show()
```
