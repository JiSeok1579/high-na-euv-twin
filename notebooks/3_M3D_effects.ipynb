# 3D Mask Effects (Phase 4)

Qualitative paper #12 six-effect stub using the reduced `src.mask_3d` interface.

```python
import os, sys
sys.path.insert(0, os.path.abspath('..'))

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401

from src.mask_3d import compare_absorber_materials
```

```python
pitches_nm = np.array([28.0, 32.0, 40.0, 64.0])
rows = []
for pitch_nm in pitches_nm:
    for summary in compare_absorber_materials(pitch_nm * 1e-9):
        rows.append((pitch_nm, summary.material.name, summary.best_focus_shift_m * 1e9, summary.contrast_loss_fraction))

for row in rows:
    print(f'pitch={row[0]:.0f} nm, material={row[1]}, BF={row[2]:.3f} nm, contrast loss={row[3]:.4f}')
```

```python
materials = sorted({row[1] for row in rows})
fig = plt.figure(figsize=(9, 5))
ax = fig.add_subplot(111, projection='3d')

for material_index, material in enumerate(materials):
    subset = [row for row in rows if row[1] == material]
    x = np.array([row[0] for row in subset])
    y = np.full_like(x, material_index, dtype=float)
    z = np.zeros_like(x)
    dz = np.array([row[2] for row in subset])
    ax.bar3d(x, y, z, 2.0, 0.4, dz, alpha=0.8)

ax.set_xlabel('pitch [nm]')
ax.set_ylabel('absorber')
ax.set_zlabel('BF shift proxy [nm]')
ax.set_yticks(range(len(materials)))
ax.set_yticklabels(materials, fontsize=8)
ax.set_title('Phase 4 Mask 3D best-focus proxy')
fig.tight_layout()
plt.show()
```
